#### **Compare their measures of central tendency**

As we know, measures of central tendency is one of the basic tools, that allow us to compare different datasets as it shows the most typical values.

##### **üåè Whole dataset**

Let's try to calculate measures of central tendency for the whole dataset first.

Calculate them for the **'tip'** column through the whole dataset and save them into the following variables:

* min => `common_tip_min`
* max => `common_tip_max`
* mean => `common_tip_mean`
* median => `common_tip_median`

```python
# YOUR CODE
common_tip_min = tips_df['tip'].min()
common_tip_max = tips_df['tip'].max()
common_tip_mean = tips_df['tip'].mean()
common_tip_median = tips_df['tip'].median()
```

Let's show the resulting values for whole dataset (we already have the code written for you üòâ)

```python
# Make a list of values
common_values = [common_tip_min, common_tip_max, common_tip_mean, common_tip_median]
# Round all the values to 4 decimal places
common_values = map(lambda x: round(x, 4), common_values)

# Make a dataframe from the list
common_mct = pd.DataFrame(common_values, index=['min', 'max', 'mean', 'median'])
# Output the dataframe
common_mct
```

### **üë®üë© Do males give more tips?**

Perform the same steps based on the column **sex**.

```python
# YOUR CODE

# 1. T√°ch d·ªØ li·ªáu th√†nh nh√≥m nam v√† n·ªØ
males_df = tips_df[tips_df['sex'] == 'Male']
females_df = tips_df[tips_df['sex'] == 'Female']

# 2. T√≠nh to√°n th·ªëng k√™ m√¥ t·∫£ cho ti·ªÅn boa c·ªßa nam gi·ªõi
males_tip_min = males_df['tip'].min()
males_tip_max = males_df['tip'].max()
males_tip_mean = males_df['tip'].mean()
males_tip_median = males_df['tip'].median()

# Hi·ªÉn th·ªã th·ªëng k√™ cho nam gi·ªõi
print("Th·ªëng k√™ ti·ªÅn boa cho nam gi·ªõi:")
males_tip_stats = pd.Series({
    'count': males_df['tip'].count(),
    'mean': males_tip_mean,
    'std': males_df['tip'].std(),
    'min': males_tip_min,
    '25%': males_df['tip'].quantile(0.25),
    '50%': males_tip_median,
    '75%': males_df['tip'].quantile(0.75),
    'max': males_tip_max
}, name='tip')
print(males_tip_stats.to_frame())
print("\n")

# 3. T√≠nh to√°n th·ªëng k√™ m√¥ t·∫£ cho ti·ªÅn boa c·ªßa n·ªØ gi·ªõi
females_tip_min = females_df['tip'].min()
females_tip_max = females_df['tip'].max()
females_tip_mean = females_df['tip'].mean()
females_tip_median = females_df['tip'].median()

# Hi·ªÉn th·ªã th·ªëng k√™ cho n·ªØ gi·ªõi
print("Th·ªëng k√™ ti·ªÅn boa cho n·ªØ gi·ªõi:")
females_tip_stats = pd.Series({
    'count': females_df['tip'].count(),
    'mean': females_tip_mean,
    'std': females_df['tip'].std(),
    'min': females_tip_min,
    '25%': females_df['tip'].quantile(0.25),
    '50%': females_tip_median,
    '75%': females_df['tip'].quantile(0.75),
    'max': females_tip_max
}, name='tip')
print(females_tip_stats.to_frame())
print("\n")

# 4. V·∫Ω bi·ªÉu ƒë·ªì histogram ƒë·ªÉ so s√°nh ph√¢n ph·ªëi
fig, axes = plt.subplots(1, 2, figsize=(20, 5)) # 1 h√†ng, 2 c·ªôt

# Bi·ªÉu ƒë·ªì 1: Ti·ªÅn boa c·ªßa nam gi·ªõi
axes[0].hist(males_df['tip'], bins=20, color='#6a0572', edgecolor='black') # M√†u t√πy ch·ªçn
axes[0].set_xlabel('Tip value')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Males tip values')
axes[0].grid(True)

# Bi·ªÉu ƒë·ªì 2: Ti·ªÅn boa c·ªßa n·ªØ gi·ªõi
axes[1].hist(females_df['tip'], bins=20, color='#ab83a1', edgecolor='black') # M√†u t√πy ch·ªçn
axes[1].set_xlabel('Tip value')
# axes[1].set_ylabel('Frequency')
axes[1].set_title('Females tip values')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
